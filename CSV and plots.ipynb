# CSV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
```

```python
target_dir = 'datasets'
wine_path = os.path.join(target_dir,'winequality-red.csv')
df = pd.read_csv(wine_path)
```

```python
df.head()
```

```python
df.to_csv('wine.csv')
```

# Scatterplot

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x ='citric acid',y ='chlorides',data = df,hue = 'quality',palette='flare')
plt.grid(True)
plt.show()
```

# Pie Chart

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,explode=[0.2,0,0,0])
plt.legend()
plt.show() 
```

# Bar Plot

```python
df['round_ph'] = np.ceil(df['pH']) 
```

```python
sns.barplot(data = df,x = 'round_ph',y  = 'alcohol')
plt.show()
```

# Subplot

```python
plt.figure(figsize=(12,9))
plt.subplot(1,2,1)
plt.scatter(x = df['pH'],y = df['alcohol'])
plt.xlabel('$pH$')
plt.ylabel('$Alcohol$')
plt.subplot(1,2,2)
plt.barh(y = df['round_ph'],width  = df['alcohol'])
plt.xlabel('$round_ph$')
plt.ylabel('$Alcohol$')
plt.grid(True)
plt.show()
```
